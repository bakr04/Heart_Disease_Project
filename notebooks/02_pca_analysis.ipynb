```python
import pandas as pd

preprocessed_data = pd.read_csv('../data/preprocessed_heart_disease.csv')
preprocessed_data
```

```python
features = preprocessed_data.drop(columns=['condition'], axis=1)
features
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=16)
pca_features = pca.fit_transform(features)
pca_features = pd.DataFrame(pca_features, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10', 'PC11', 'PC12','PC13', 'PC14', 'PC15', 'PC16'])
pca_features 
```

```python
pca_variance = pca.explained_variance_ratio_
print(pca_variance)
```

```python
pca_processed_data = pd.concat([pca_features, preprocessed_data[['condition']]], axis=1) 
pca_processed_data
```

```python
import numpy as np
import matplotlib.pyplot as plt

cumulative_variance = np.cumsum(pca_variance)

optimal_components = np.argmax(cumulative_variance >= 0.98) + 1

print("Explained variance ratio per component:", pca_variance)
print("Cumulative explained variance:", cumulative_variance)
print("Optimal number of components (98% variance):", optimal_components)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(pca_features['PC1'], pca_features['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs Number of Components')
plt.axhline(y=0.98, color='r', linestyle='--', label='98% threshold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
pca_processed_data.to_csv('../data/pca-processed_heart_disease.csv', index=False)
```

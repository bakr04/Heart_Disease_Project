```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import joblib

final_data = pd.read_csv('../data/final_processing_afterfeatureselection.csv')
final_data
```

```python
x = final_data.drop(columns=['condition'])
y = final_data['condition']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True, stratify=y)
```

```python

from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import f1_score
import warnings
from sklearn.exceptions import ConvergenceWarning
import joblib

warnings.filterwarnings("ignore", category=ConvergenceWarning)


baseline_models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "LinearSVC": LinearSVC(random_state=42)
}


param_grids = {
    "Logistic Regression": {
        "C": [0.01,0.1,1,10,100],
        "penalty": ["l1","l2"],
        "solver": ["saga", "liblinear"],
        "fit_intercept": [True, False],
        "max_iter" : [1000, 2000, 3000, 4000, 5000]
    },
    "Decision Tree": {
        "max_depth": [None, 5, 10, 20],
        "min_samples_split": [2,5,10],
        "min_samples_leaf": [1,2,4],
        "criterion": ["gini", "entropy"]
    },
    "Random Forest": {
        "n_estimators": [100,200,300,500,700],
        "max_depth": [None,10,20],
        "min_samples_split": [2,5],
        "min_samples_leaf": [1,2],
        "max_features": ["sqrt","log2", None]
    },
    "SVM": {
        "C": [0.1,1,10],
        "kernel": ["linear", "rbf", "poly"],
        "gamma": ["scale", "auto"]
    },
    "LinearSVC": {
        "C": [0.1,1,10],
        "loss": ["hinge","squared_hinge"],
        "multi_class": ["ovr","crammer_singer"],
        "max_iter" : [1000, 2000, 3000, 4000, 5000]

    }
}

cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)


print("📊 Baseline Performance")
baseline_results = {}

for name, model in baseline_models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    f1 = f1_score(y_test, y_pred, average="weighted")
    baseline_results[name] = f1
    print(f"➡️ {name}: F1 Score = {f1:.4f}")

print("="*60)

best_models = {}
optimized_results = {}

for name, model in baseline_models.items():
    print(f"\n🔎 Optimizing {name}...")

    grid = param_grids[name]
    total_combinations = np.prod([len(v) for v in grid.values()])
    n_iter = min(30, total_combinations) 

    search = RandomizedSearchCV(
        model, 
        param_distributions=grid, 
        n_iter=n_iter, 
        cv=cv_strategy,
        scoring="f1_weighted",
        n_jobs=1, 
        random_state=42
    )
    search.fit(x_train, y_train)
    best_model = search.best_estimator_

    best_models[name] = best_model
    y_pred = best_model.predict(x_test)
    f1 = f1_score(y_test, y_pred, average="weighted")
    optimized_results[name] = f1

    print(f"✅ {name} optimized")
    print(f"   CV Best F1 Score: {search.best_score_:.4f}")
    print(f"   Test F1 Score: {f1:.4f}")
    print(f"   Best Params: { {k:v for k,v in search.best_params_.items()} }")
    
    joblib.dump(best_model, f"../models/{name.replace(' ','_')}_optimized.pkl")


print("\n📊 Performance Comparison (Baseline vs Optimized)")
comparison = pd.DataFrame({
    "Baseline F1": baseline_results,
    "Optimized F1": optimized_results
})
print(comparison)
```
